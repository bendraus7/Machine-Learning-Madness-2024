```python
base_urls=[
          'https://web.archive.org/web/20110311233233/http://www.kenpom.com/',#2011
          'https://web.archive.org/web/20120311165019/http://kenpom.com/',#2012
          'https://web.archive.org/web/20130318221134/http://kenpom.com/',#2013
          'https://web.archive.org/web/20140318100454/http://kenpom.com/',#2014
          'https://web.archive.org/web/20150316212936/http://kenpom.com/',#2015
          'https://web.archive.org/web/20160314134726/http://kenpom.com/',#2016
          'https://web.archive.org/web/20170314003320/http://kenpom.com/',#2017
          'https://web.archive.org/web/20180312133836/https://kenpom.com/',#2018
          'https://web.archive.org/web/20190317205307/https://kenpom.com/',#2019
          # Tournament wasn't played on 2020
          'https://web.archive.org/web/20210315132114/https://kenpom.com/',#2021
          'https://web.archive.org/web/20220315152153/https://kenpom.com/',#2022
          'https://web.archive.org/web/20230314135153/https://kenpom.com/', #2023
          'https://web.archive.org/web/20240318160959/https://kenpom.com/', #2024
          'https://kenpom.com/'
         ]

base_urls2=[
          'https://kenpom.com/index.php?y=2011',#2011
          'https://kenpom.com/index.php?y=2012',#2012
          'https://kenpom.com/index.php?y=2013',#2013
          'https://kenpom.com/index.php?y=2014',#2014
          'https://kenpom.com/index.php?y=2015',#2015
          'https://kenpom.com/index.php?y=2016',#2016
          'https://kenpom.com/index.php?y=2017',#2017
          'https://kenpom.com/index.php?y=2018',#2018
          'https://kenpom.com/index.php?y=2019',#2019
          # Tournament wasn't played on 2020
          'https://kenpom.com/index.php?y=2021',#2021
          'https://kenpom.com/index.php?y=2022',#2022
          'https://kenpom.com/index.php?y=2023', #2023
          'https://kenpom.com/index.php?y=2024', #2024
          'https://kenpom.com/'
         ]


years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2021, 2022, 2023, 2024]

def scrape_archive(url,year):
    """
    Imports raw data from a kenpom archive into a dataframe
    """
    
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    table_full = soup.find_all('table', {'id': 'ratings-table'})

    thead = table_full[0].find_all('thead')
    table = table_full[0]
    
    for weird in thead:
        table = str(table).replace(str(weird), '')

    df = pd.read_html(table)[0]
    df['year'] = year
    
    return df
```

```python
import time  # Import the time module

def scraping(df, years):
    for url, year in zip(base_urls, years):
        print(year)
        print(f'Scraping: {url}')
        
        # Scrape the archive for the current year
        archive = scrape_archive(url, year)
        
        # Concatenate the scraped data to the existing DataFrame
        df = pd.concat((df, archive), axis=0)
        
        # Increment the year
        year += 1
        
        # Wait for 10 seconds before scraping the next page
#         print("Waiting for 10 seconds before scraping the next page...")
#         time.sleep(10)  # Pause execution for 10 seconds
        
    # Rename the columns of the DataFrame
    df.columns = ['Rank', 'Team', 'Conference', 'W-L', 'AdjEM', 
                  'AdjustO', 'AdjustO Rank', 'AdjustD', 'AdjustD Rank',
                  'AdjustT', 'AdjustT Rank', 'Luck', 'Luck Rank', 
                  'SOS AdjEM', 'SOS AdjEM Rank', 'SOS OppO', 'SOS OppO Rank',
                  'SOS OppD', 'SOS OppD Rank', 'NCSOS AdjEM', 'NCSOS AdjEM Rank', 'Year']
    
    # You may want to uncomment and modify this line if necessary
    # df.TeamName = df.TeamName.apply(lambda x: re.sub('\d', '', x).strip()).replace('.','')
    
    return df
```

```python
# Lambda that returns true if given string is a number and a valid seed number (1-16)
valid_seed = lambda x: True if str(x).replace(' ', '').isdigit() \
                and int(x) > 0 and int(x) <= 16 else False

# Use lambda to parse out seed/team
result['Seed'] = result['Team'].apply(lambda x: x[-2:].replace(' ', '') \
                              if valid_seed(x[-2:]) else np.nan )

result['Team'] = result['Team'].apply(lambda x: x[:-2] if valid_seed(x[-2:]) else x)
```

```python
# Split W-L column into wins and losses
result['Wins'] = result['W-L'].apply(lambda x: int(re.sub('-.*', '', x)) )
result['Losses'] = result['W-L'].apply(lambda x: int(re.sub('.*-', '', x)) )
result.drop('W-L', inplace=True, axis=1)
```
